```python
import pandas
import matplotlib.pyplot as plt
import matplotlib
from numpy import *
```

```python
fr1 = pandas.read_csv('alarm(1).csv')
```

```python
nl=fr1['NE_LABEL']
title=fr1['ALARM_TITLE_TEXT']
```

```python
nls=[]
for n_l in nl:
    if n_l not in nls:
        nls.append(n_l)
nll={}
for i in range(37237):
    for n_l in nls:
        if n_l==nl[i]:
            if n_l in nll:
                nll[n_l]+=1
            else:
                nll[n_l]=1
```

```python
def flit(n):
    nlt={}
    for d in nll:
        if nll[d]>n:
            nlt[d]=nll[d]
    return nlt
```

```python
nlt=flit(30)
nlt
```

```python
def nlgetid(nlt):
    nld={}
    for n in nlt:
        nld[n]=[]
        for i in range(37237):
            if nl[i]==n:
                if nl[i] in nld:
                    nld[n].append(i)
                else:
                    nld[n]=[i]
    return nld
```

```python
nld=nlgetid(nlt)
```

```python
nld
```

```python
et=fr1['EVENT_TIME']
```

```python
def gt1(t1):
    gt={}
    for ts in nld[t1]:
        if int(et[ts][-5:-3]) in gt:
            gt[int(et[ts][-5:-3])]+=1
        else:
            gt[int(et[ts][-5:-3])]=1
    return gt
```

```python
def gt2(t1):
    gt={}
    for ts in nld[t1]:
        if int(et[ts][-2:]) in gt:
            gt[int(et[ts][-2:])]+=1
        else:
            gt[int(et[ts][-2:])]=1
    return gt
```

```python
from matplotlib.font_manager import FontProperties
font_set = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=12)
def plot(t1,function,x):
    plt.figure()
    plt.bar(list(function(t1).keys()),list(function(t1).values()),align =  'center')
    plt.axis([-1,x,0,max(list(function(t1).values()))*1.2])
    plt.title(t1,fontproperties=font_set)
    plt.show()
```

```python
for i in nlt:
    plot(i,gt1,24)
```

```python
for i in nlt:
    plot(i,gt2,60)
```
