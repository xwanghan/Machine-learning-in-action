```python
import re
def gd(line):
    searchObj = re.search( r'(.*?)/(.*?)/(.*?) ', line)
    msd1={1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
    msd2={1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
    if searchObj.group(1)=='2019':
        for ms in msd1:
            if int(searchObj.group(2))==ms:
                days=int(searchObj.group(3))-1
                for m in range(1,ms):
                    days+=msd1[m]
                return days*24*60+int(line[-5:-3])*60+int(line[-2:])
    if searchObj.group(1)=='2020':
        for ms in msd2:
            if int(searchObj.group(2))==ms:
                days=int(searchObj.group(3))-1
                for m in range(1,ms):
                    days+=msd2[m]
                return 365*24*60+days*24*60+int(line[-5:-3])*60+int(line[-2:])
```

```python
import sys

class ProgressBar():

    def __init__(self, max_steps):
        self.max_steps = max_steps
        self.current_step = 0
        self.progress_width = 50

    def update(self, step=None):
        self.current_step = step

        num_pass = int(self.current_step * self.progress_width / self.max_steps) + 1
        num_rest = self.progress_width - num_pass 
        percent = (self.current_step+1) * 100.0 / self.max_steps 
        progress_bar = '[' + '■' * (num_pass-1) + '▶' + '-' * num_rest + ']'
        progress_bar += '%.2f' % percent + '%' 
        if self.current_step < self.max_steps - 1:
            progress_bar += '\r' 
        else:
            progress_bar += '\n' 
        sys.stdout.write(progress_bar) 
        sys.stdout.flush()
        if self.current_step >= self.max_steps:
            self.current_step = 0
            print
```

```python
class NeuralNetwork():
    def __init__(self): 
        #converting weights to a 3 by 1 matrix with values from -1 to 1 and mean of 0
        self.synaptic_weights1 = 2 * np.random.random((3, 2)) - 1
        self.synaptic_weights2 = 2 * np.random.random((2, 1)) - 1
    def sigmoid(self, x):
        #applying the sigmoid function
        return 1 / (1 + np.exp(-x))
    def sigmoid_derivative(self, x):
        #computing derivative to the Sigmoid function
        return x * (1 - x)
    def train(self, training_inputs, training_outputs,training_iterations,learning_rate):
        #training the model to make accurate predictions while adjusting weights continually
        max_batchs = 300
        progress_bar = ProgressBar(training_iterations)
        for iteration in range(training_iterations):
            #siphon the training data via  the neuron
            a1,output = self.think(training_inputs)
            #computing error rate for back-propagation
            error = training_outputs - output
            #performing weight adjustments
            adjustments2 = np.dot(a1.T, error * self.sigmoid_derivative(output))
            self.synaptic_weights2 += adjustments2*learning_rate
            adjustments1 = np.dot(training_inputs.T, error * self.sigmoid_derivative(a1))
            self.synaptic_weights1 += adjustments1*learning_rate
            progress_bar.update(iteration)
    def think(self, inputs):
        #passing the inputs via the neuron to get output   
        #converting values to floats
        inputs = inputs.astype(float)
        a1 = self.sigmoid(np.dot(inputs, self.synaptic_weights1))
        output = self.sigmoid(np.dot(a1, self.synaptic_weights2))
        return a1,output
```

```python
if __name__ == "__main__":
    #initializing the neuron class
    neural_network = NeuralNetwork()
    print("Beginning Randomly Generated Weights1: ")
    print(neural_network.synaptic_weights1)
    print("Beginning Randomly Generated Weights2: ")
    print(neural_network.synaptic_weights2)
    #training data consisting of 4 examples--3 input values and 1 output
    training_inputs = np.array(inputofnn)
    training_outputs = np.array(outputofnn)/4
    #training taking place
    neural_network.train(training_inputs, training_outputs, 10000,1)
    print("Ending Weights1 After Training: ")
    print(neural_network.synaptic_weights1)
    print("Ending Weights2 After Training: ")
    print(neural_network.synaptic_weights2)
```
