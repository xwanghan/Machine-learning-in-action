```python
title=fr1['ALARM_TITLE_TEXT']
et=fr1['EVENT_TIME']
t=[]
for tit in title:
    if tit not in t:
        t.append(tit)
tl={}
for i in range(len(title)):
    for tit in t:
        if tit==title[i]:
            if tit in tl:
                tl[tit]+=1
            else:
                tl[tit]=1
```

```python
nl=fr1['NE_LABEL']
```

```python
nls=[]
for n_l in nl:
    if n_l not in nls:
        nls.append(n_l)
nll={}
for i in range(37237):
    for n_l in nls:
        if n_l==nl[i]:
            if n_l in nll:
                nll[n_l]+=1
            else:
                nll[n_l]=1
```

```python
def nlgetid(nlt):
    nld={}
    for n in nlt:
        nld[n]={}
        for i in range(37237):
            if nl[i]==n:
                if nl[i] in nld:
                    if title[i] in nld[n]:
                        nld[n][title[i]]+=1
                    else:
                        nld[n][title[i]]=1
                else:
                    nld[n][title[i]]=1

    return nld
```

```python
nld=nlgetid(nll)
```

```python
nld
```

```python
don={}
for i in range(37237):
    for tit in nls:
        if tit==nl[i]:
            if tit not in don:
                don[tit]=[i]
            else:
                don[tit].append(i)
```

```python
don
```

```python
import re
def gm(line):
    searchObj = re.search( r'(.*?)/(.*?)/', line)
    if searchObj.group(1)=='2019':
        return int(searchObj.group(2))
    if searchObj.group(1)=='2020':
        return int(searchObj.group(2))+12
```

```python
def gt(t1):
    gt={}
    for ts in don[t1]:
        if int(gm(et[ts])) in gt:
            gt[int(gm(et[ts]))]+=1
        else:
            gt[int(gm(et[ts]))]=1
    return gt
```

```python
from matplotlib.font_manager import FontProperties
font_set = FontProperties(fname=r"c:\windows\fonts\simsun.ttc", size=12)
def plot(t1):
    plt.figure()
    plt.bar(list(gt(t1).keys()),list(gt(t1).values()),align =  'center')
    plt.axis([-1,17,0,max(list(gt(t1).values()))*1.2])
    plt.title(t1,fontproperties=font_set)
    plt.show()
```

```python
for i in nls:
    plot(i)
```

```python
newnld={}
for n in nld:
    newnld[n]={}
    for x in nld[n]:
        newnld[n][x]=(nld[n][x]/tl[x])*100
```

```python
perarray={}
for n in newnld:
    perarray[n]=[]
    for s in t:
        nlwitht=[]
        for x in newnld[n]:
            nlwitht.append(x)
            if s==x:
                perarray[n].append(newnld[n][x])
        if s not in nlwitht:
            perarray[n].append(0)
```
