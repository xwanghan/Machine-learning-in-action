```python
import re
def gd(line):
    searchObj = re.search( r'(.*?)/(.*?)/(.*?) ', line)
    msd1={1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
    msd2={1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
    if searchObj.group(1)=='2019':
        for ms in msd1:
            if int(searchObj.group(2))==ms:
                days=int(searchObj.group(3))-1
                for m in range(1,ms):
                    days+=msd1[m]
                return days*24*60+int(line[-5:-3])*60+int(line[-2:])
    if searchObj.group(1)=='2020':
        for ms in msd2:
            if int(searchObj.group(2))==ms:
                days=int(searchObj.group(3))-1
                for m in range(1,ms):
                    days+=msd2[m]
                return 365*24*60+days*24*60+int(line[-5:-3])*60+int(line[-2:])
```

```python
def gety(sets,tandnl):
    y=[]
    for labels in tandnl:
        yl=[]
        for keys in sets:
            try:
                yl.append(dohast[keys][labels[0]][labels[1]])
            except KeyError:
                yl.append(0)
        y.append(yl)
    y=mat(y)
    return y
```

```python
def gtandnl(sets):
    tandnl=[]
    for keys in sets:
        for keys1 in dohast[keys]:
            for keys2 in dohast[keys][keys1]:
                if [keys1,keys2] not in tandnl:
                    tandnl.append([keys1,keys2])
    return tandnl
```

```python
def plot1(sets,tandnl):
    y=gety(sets,tandnl)
    hzb=[]
    zzb=[]
    for i in range(shape(y)[0]):
        for j in range(shape(y)[1]):
            if y[i,j]>0:
                hzb.append(i)
                zzb.append(j)
    plt.scatter(zzb,hzb,marker='.')
    plt.yticks(list(arange(shape(y)[0])),tandnl)
    plt.grid()
    plt.show()
```
