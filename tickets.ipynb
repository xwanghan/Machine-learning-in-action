```python
import pandas
import matplotlib.pyplot as plt
from numpy import *
```

```python
def datamatrix(filename):
    fr = pandas.read_csv(filename)
    company = fr['MAINEQUIPMENTFACTORYNAME']
    time1 = fr['AFFACTBUSIDURATION']
    time2=fr['DEALDURATION']
    r1=fr['LINKFAULTREASONSORT']
    r2=fr['LINKFAULTREASONSUBSECTION']
    r3=fr['LINKFAULTREASONTHIRD']
    s=fr['LINKDEALSTEP']
    ms1=fr['MAINNETSORTONE']
    ms2=fr['MAINNETSORTTWO']
    ms3=fr['MAINNETSORTTHREE']
    lenth=len(fr['LINKDEALSTEP'])
    return  company,time1,time2,r1,r2,r3,s,lenth,ms1,ms2,ms3
```

```python
c=[]
for company in n:
    if company not in c:
        c.append(company)
cl=len(c)
c ,cl
```

```python
def getidofc(a):
    idofc=[]
    id=0
    for company in n:
        if company == c[a]:
            idofc.append(id)
        id+=1
    return idofc
def getT(a):
    t1a=[]
    t2a=[]
    for id in a:
        t1a.append(t1[id])
        t2a.append(t2[id])
    return t1a,t2a
def plot(t1a,t2a):
    ax = plt.subplot(111)
    ax.scatter(t1a, t2a)
    plt.show()
```

```python
for x in range(cl):
    a0=getidofc(x)
    t1a0,t2a0=getT(a0)
    plot(t1a0,t2a0)
```

```python
def filter(n):
    f={}
    for key1 in tree3:
        f[key1]={}
        for key2 in tree3[key1]:
            f[key1][key2]={}
            for key3 in tree3[key1][key2]:
                f[key1][key2][key3]={}
                for key4 in tree3[key1][key2][key3]:
                    f[key1][key2][key3][key4]={}
                    for key5 in tree3[key1][key2][key3][key4]:
                        f[key1][key2][key3][key4][key5]={}
                        for key6 in tree3[key1][key2][key3][key4][key5]:
                            f[key1][key2][key3][key4][key5][key6]={}
                            for key7 in tree3[key1][key2][key3][key4][key5][key6]:
                                f[key1][key2][key3][key4][key5][key6][key7]=0
                                if tree3[key1][key2][key3][key4][key5][key6][key7] >n:
                                    f[key1][key2][key3][key4][key5][key6][key7]=tree3[key1][key2][key3][key4][key5][key6][key7]
                                else:
                                    del  f[key1][key2][key3][key4][key5][key6][key7]
                            if f[key1][key2][key3][key4][key5][key6]=={}:
                                del f[key1][key2][key3][key4][key5][key6]
                        if f[key1][key2][key3][key4][key5]=={}:
                            del f[key1][key2][key3][key4][key5]
                    if f[key1][key2][key3][key4]=={}:
                        del f[key1][key2][key3][key4]
                if f[key1][key2][key3]=={}:
                    del f[key1][key2][key3]
            if f[key1][key2]=={}:
                del f[key1][key2]
        if f[key1]=={}:
            del f[key1]
    return f
```

```python
def getline(ls):
    x=[]
    y=[]
    for s in ls:
        x.append([1,t1[s]])
        y.append(t2[s])
    x=mat(x)
    y=mat(y)
    y=y.T
    xTx=x.T*x
    w=xTx.I*(x.T*y)
    return w
def plotline(t1a,t2a,w,x1,x2):
    ax = plt.subplot(111)
    ax.scatter(t1a, t2a)
    x=linspace(x1,x2)
    y=x*w[1]+w[0]
    plt.plot(x,y)
    plt.show()
```
